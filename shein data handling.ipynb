```python
all_materials = []
```

```python
def get_materials(material_row):
    materials_dict = {}
    if material_row:
        for material in material_row.split(','):
            materials_dict[material.split('%')[1].strip()] = float(material.split('%')[0])
    all_materials.append(materials_dict)
```
